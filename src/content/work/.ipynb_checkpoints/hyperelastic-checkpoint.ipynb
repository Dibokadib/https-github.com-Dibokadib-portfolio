```python
from fenics import *
parameters.plotting_backend = 'matplotlib'
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
L2 = 0.12
L1 = L2/10
P1 = Point(0.,0.)
P2 = Point(L1,L2)
nx = 4
ny = nx * int(L2/L1)
```

```python
mesh = RectangleMesh(P1,P2,nx,ny)
imperfection = Expression(("4*x[1]/L2*(1-x[1]/L2)*a","0"),a=0.1*L1,L2=L2,degree=2)
ALE.move(mesh,imperfection)
plot(mesh)
```

```python
element = VectorElement("CG", triangle, 1)
V = FunctionSpace(mesh, element)
u = Function(V)
ut = TestFunction(V)
delta_u = TrialFunction(V)
```

```python
E, nu = 1.3e+6, 0.45
lmbda = Constant(E*nu/((1+nu)*(1-2*nu)))
mu = Constant(E/(2+2*nu))
Id = Identity(2)
F = Id + grad(u)
C = F.T*F
E = 1./2.*(C-Id)
J = det(F)
W1 = lmbda/2.*tr(E)**2 + mu*inner(E,E)
W2 = mu/2*(tr(C)-2)-mu*ln(J)+lmbda/2.*ln(J)**2
energy = W2*dx
```

```python
energy_u = derivative(energy,u,ut)
energy_uu = derivative(energy_u,u,delta_u)
```

```python
bc1 = DirichletBC(V, Constant((0.,0.)), "x[1]<1.e-8 && on_boundary")
uL = Expression(("0.","-t"),t=0.01)
bc2 = DirichletBC(V, uL, "x[1]>(%s-1.e-8) && on_boundary"%L2)
bcs = [bc1, bc2]
```

```python
u = interpolate(Constant((0.,0.)),V)
problem = NonlinearVariationalProblem(energy_u, u, bcs=bcs, J=energy_uu)
solver = NonlinearVariationalSolver(problem)
```

```python
#u = interpolate(Constant((0.,0.)),V)
loadings = np.linspace(0,L2/5.,101)
out = loadings*0
fileu = File("disp.pvd")
u.rename("u","u")
# print(loadings)
for (i,load) in enumerate(loadings):
    uL.t = load
    solver.solve()
    outdisp = u(L1/2.,L2/2.)
    out[i] = outdisp[0]
    fileu << (u, load)
```

```python
plot(u,mode="displacement")
```
